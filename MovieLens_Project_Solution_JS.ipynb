### 8. Visualize how popularity of genres has changed over the years

### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 

```python
rating_sub1 = rating.groupby('movie id').count().reset_index().copy() #Group by movie id then get rating count per movie
rating_series2 = rating.groupby('movie id').sum()['rating'].reset_index()  #Group by movie id then get rating sum per movie
rating_sub1['Rating Sum'] = rating_series2['rating'] #Add rating sum to the rating subset dataframe
rating_series3 = rating_sub1['Rating Sum'].div(rating_sub1['rating'], axis= 0)  #Calculate avg rating per movie
rating_sub1['Rating Avg'] = rating_series3 #Add average rating (rating sum/rating count) to the rating subset dataframe
rating_sub1.drop(['user id','timestamp'], axis=1, inplace=True) #Drop user id and timestamp column since it is not needed
movie_sub3 = movie.copy() #Make a copy of movie dataframe
movie_sub3 = movie_sub3.merge(rating_sub1) #Merge rating count, rating sum and rating avg to movie subset dataframe per movie id
movie_sub3.sort_values(by=['Rating Avg'],ascending=False, inplace=True) #Sort movie based on rating avg in descending order
movie_sub3[movie_sub3['rating'] > 100].head(25) #Get Top 25 high rated movies with rating count greater than 100
```

### 10. See gender distribution across different genres check for the validity of the below statements

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women

**compare the percentages**


```python
merged_df = movie.copy() #Create copy of movie dataframe
merged_df = merged_df.merge(rating) #Merge rating dataframe with the movie dataframe 
merged_df = merged_df.merge(user) #Merge user dataframe with the movie dataframe
merged_df.head()
```

```python
merged_df.groupby('gender')['rating'].count() #Get rating count gender wise in total
```

```python
merged_df.groupby('gender').sum() #Get rating sum gender wise across genre
```

```python
merged_df.groupby('gender').sum().loc[['M','F'],['Drama','Romance','Sci-Fi']] #Fetch rating sum gender wise for Drama, Romance and Sci-Fi genre
```

```python
#Get percentage rating gender wise for Drama, Romance and Sci-Fi genre
merged_df.groupby('gender').sum().loc[['M','F'],['Drama','Romance','Sci-Fi']].div(merged_df.groupby('gender')['rating'].count(), axis= 0)*100 
```
